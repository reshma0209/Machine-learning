Find average SOLD PRICE for each age category, group all records by AGE and then apply mean() on SOLD PRICE column

Perform joining of data frame- compare the average SOLD PRICE for different AGE categories with the different age and PLAYING ROLE categories, we need to merge the DataFrames soldprice_by_age and soldprice_by_ age_role. The common column is AGE and this needs outer join. Age is considered as categorical varaibale(as 1,2,3)

```python
#Perform joining of data frame
soldprice_by_age=df[['AGE','SOLD PRICE']]
soldprice_by_age_role=df[['AGE','PLAYING ROLE','SOLD PRICE']]
categories = [1, 2, 3]
soldprice_by_age.loc[:, 'AGE'] = pd.Categorical(soldprice_by_age['AGE'], categories=categories)
soldprice_by_age_role.loc[:, 'AGE'] = pd.Categorical(soldprice_by_age_role['AGE'], categories=categories)
pd.merge(soldprice_by_age,soldprice_by_age_role,on='AGE',how='outer')
```

To find whether players carry a premium if they belong to a specific AGE and PLAYING ROLE category. The premium (we will call it change) is calculated in percentage terms and calculated

```python
#To find whether players carry a Premium
avg1=df.groupby("AGE")["SOLD PRICE"].mean()
avg2=df.groupby(["AGE","PLAYING ROLE"])["SOLD PRICE"].mean()
#Premium (we will call it change)
change=(avg2-avg1)/avg1
change
```
